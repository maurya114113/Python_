```python
import matplotlib.pyplot as plt
from matplotlib import style
age_men=25,11,68,18,27,28,15,43,58,63,43,65,51,36,33,26,23,35,49,58
plt.hist(age_men,bins=10,edgecolor='black',color='blue')
style.use('ggplot')
plt.xlabel("age of men",fontsize=15)
plt.ylabel("age of women",fontsize=15)
plt.title("Histogram plot of age of men",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
age_men=25,11,68,18,27,28,15,43,58,63,43,65,51,36,33,26,23,35,49,58
age_women=14,6,2,2,4,15,11,18,1,7,2,1,3,4,10,2
plt.hist([age_men,age_women],bins=10,edgecolor='black',color=['blue','green'])
style.use('ggplot')
plt.xlabel("age of men and women",fontsize=15)
plt.ylabel("frequeny of age ",fontsize=15)
plt.title("Histogram plot of age of men and women",fontsize=15)
plt.show()
```
